```python
import pandas as pd
import numpy as np
import matplotlib
```

```python
#rolling two dice and adding the values ​​- Random variables
die = pd.DataFrame([1,2,3,4,5,6])
sum_of_dice = die.sample(2, replace=True).sum().loc[0]
print('A soma dos dados é: ',sum_of_dice)


```

```python
#Throwing the dice N times and adding up the valuesn = 50
n= 5
result = [die.sample(2, replace= True).sum().loc[0] for i in range(n)]
print(result)
```

```python
freq = pd.DataFrame(result)[0].value_counts() #count Values
sort_freq = freq.sort_index() #sort index
sort_freq
sort_freq.plot(kind ='bar', color ='blue')
```

```python
#frequencia relativa = frequencia / numero de tentativas
relative_freq = sort_freq / n
relative_freq.plot(kind= 'bar', color = 'blue')
```
